```python
from __future__ import print_function
import numpy as np
import pylab as pl
#import seaborn
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

```python
mymean = 100
```

## Chi-Squared

```python
distributions = ['chisq']
df = mymean
mydictionary = {}
mydictionary['chisq'] = np.random.chisquare(df,size=100)
pl.hist(mydictionary['chisq'],bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

### Figure 1: Chi Squared distribution: df = 100, N = 100

```python
print ('Chi-Squared mean: %.2f \nStandard Deviation: %.2f'\
      %(mydictionary['chisq'].mean(), mydictionary['chisq'].std()))
```

```python
#mysize = 2000 / (np.array(range(1,100) + [10]))
#mysize = (2000 / (np.array(range(1,101)))).astype(int)
#mysize = (np.random.rand(100)*1000).astype(int)
mysize = np.random.random_integers(10,2000,100)
#print(mysize, mysize.shape, mysize.mean())
```

```python
mydictionary = {}
mydictionary['chisq'] = {}
for n in mysize: 
    mydictionary['chisq'][n] = np.random.chisquare(df, size = n)
```

```python
mydictionary['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in mydictionary['chisq']:
    if not type(nn) == str: 
        mydictionary['chisq']['means'][nn] = mydictionary['chisq'][nn].mean()
        
        axchisq_mu_n.plot(nn, mydictionary['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi-Squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df,df], 'k')
```

```python
allmeans = list(mydictionary['chisq']['means'].values())
pl.figure(figsize=(10,6))
pl.hist(allmeans, bins=30)
pl.xlabel('sample mean', fontsize=18)
pl.ylabel('N', fontsize=18)
```
